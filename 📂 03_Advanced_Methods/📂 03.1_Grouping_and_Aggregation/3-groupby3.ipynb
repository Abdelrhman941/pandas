```python
df = pd.read_csv("../../datasets/Dummy Sales_Data_v1.csv")
df.head()
```

--- 
> ### GROUP SIZES
> üëâ **`size()`** : to get the count of each group

```python
df.groupby("Product_Category").size()
```

--- 
> ### GET FIRST AND LAST
> üëâ These functions return the first and last records after data is split into different groups. Rather than referencing the index, it gives out the first or last row appearing in all the groups.

```python
df.groupby("Product_Category").first()
```

```python
df.groupby("Product_Category").last()
```

---
> ### get_group
> üëâ used to select or extract only one group from the GroupBy object

```python
df.head()
```

```python
df.groupby("Product_Category").get_group("Office")
```

```python
# Select only required columns
df_columns = df.groupby("Product_Category") [["UnitPrice(USD)","Quantity"]].mean()
df_columns
```

```python
# What orders belongs to category Home and with quantity less than 50?
s=df.groupby("Product_Category").get_group("Home")
type(s)     # pandas.core.frame.DataFrame
```

```python
# print all Quantity value  which is _Home_ that less than 50 
s[s["Quantity"]<50]
```
