```python
data2=pd.read_csv("../../datasets/Flavors.csv")
data2
```

```python
# Which Base Flavor is better on average ?
data2.groupby("Base Flavor").mean(numeric_only=True)
```

```python
# Count Ice creams based on Base Flavor ?
data2.groupby("Base Flavor").count()
```

```python
# Return the worst Ice Cream in Texture Rating for each category(Base Flavor)
print(data2.groupby("Base Flavor").min())   # not good to get min ice-cream
print("---------")
result=data2.groupby("Base Flavor")["Texture Rating"].min()
print(result)
print("---------")
data2[np.isin(data2["Texture Rating"],result.values)]
```

```python
data2.groupby("Base Flavor")["Texture Rating"].agg(["mean","min","max","count","sum","median"])
```

--- 
> ### Groupby multiple columns

```python
data2.groupby(["Base Flavor"]).mean(numeric_only=True)
```

```python
data2.groupby(["Base Flavor","Liked"]).mean(numeric_only=True)
```

```python
data2.groupby(["Base Flavor","Liked"]).agg({"Texture Rating":["min","max","mean"]})
```
