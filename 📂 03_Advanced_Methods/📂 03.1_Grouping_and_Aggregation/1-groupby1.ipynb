## üìä **Difference Between `mean()` & `groupby().mean()`**  

### üîé **1Ô∏è‚É£ Overall Mean ‚Äì `mean()`**  
‚úîÔ∏è Computes the **average** of a specific column across **all data**  

```python
print(data.beer_servings.mean())  # Average beer_servings for all countries
```

### üåç **2Ô∏è‚É£ Grouped Mean ‚Äì `groupby().mean()`**
‚úîÔ∏è **Calculates the mean for each group separately**
‚úîÔ∏è Useful when analyzing data by categories (e.g., continents)

```python
print(data.groupby("continent").beer_servings.mean())  # Average beer_servings per continent
```

### ‚ö†Ô∏è **Handling Errors ‚Äì `numeric_only=True`**
‚úîÔ∏è groupby().mean() _sometimes fails if non-numeric columns exist_
‚úîÔ∏è Use numeric_only=True to avoid errors

```python
data.groupby("continent").mean(numeric_only=True)
```

> ## üöÄ Summary:

`mean()` ‚Üí **Calculates a single overall average**        
`groupby().mean()` ‚Üí **Computes averages per category/group**

````python
data.head()
````

````python
print(data.beer_servings.mean())                            # Average beer_servings for all countries
print("-----------------")
print(data.groupby("continent").beer_servings.mean())       # Average beer_servings per continent
print("-----------------")
# data.groupby("continent").mean()  # sometime it error
data.groupby("continent").mean(numeric_only=True)
````

````python
data.groupby("continent").mean(numeric_only=True).plot(kind="bar")
plt.show()
````

````python
# What is the maximum beer_servings for each continent?
result=data.groupby("continent").beer_servings.max()
print(result)

# Return rows with the maximum beer_servings in each continent
data[np.isin(data.beer_servings, result.values)]
````
